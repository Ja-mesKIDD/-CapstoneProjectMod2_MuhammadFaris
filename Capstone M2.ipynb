## **BACKGROUND**

### **New York City TLC Trip Record Data Background**


The New York City Taxi and Limousine Commission (TLC) Trip Record dataset provides a detailed record of taxi trips taken in New York City. This dataset is a valuable resource for analyzing and understanding the dynamics of taxi operations, passenger behavior, and the overall transportation landscape in the city. Each row in the dataset represents a single taxi trip, and the columns provide information about various aspects of that trip.

**Data Columns and Descriptions:**

1. **VendorID:**
   - *Description:* A numeric code indicating the taxi service provider (vendor) associated with the trip.

2. **lpep_pickup_datetime:**
   - *Description:* The date and time when the taxi picked up the passenger.

3. **lpep_dropoff_datetime:**
   - *Description:* The date and time when the taxi dropped off the passenger.

4. **store_and_fwd_flag:**
   - *Description:* A flag indicating whether the trip data was stored in the vehicle's memory (and forwarded later) before being transmitted to the system.

5. **RatecodeID:**
   - *Description:* A numeric code indicating the final rate code in effect at the end of the trip, representing the fare calculation method.

6. **PULocationID:**
   - *Description:* The TLC Taxi Zone code for the pickup location.

7. **DOLocationID:**
   - *Description:* The TLC Taxi Zone code for the drop-off location.

8. **passenger_count:**
   - *Description:* The number of passengers in the taxi during the trip.

9. **trip_distance:**
   - *Description:* The distance (in miles) traveled during the trip.

10. **fare_amount:**
    - *Description:* The base fare amount charged for the trip.

11. **extra:**
    - *Description:* Additional charges beyond the base fare, such as peak-hour charges.

12. **mta_tax:**
    - *Description:* The Metropolitan Transportation Authority (MTA) tax, a fixed fee applied to each taxi trip.

13. **tip_amount:**
    - *Description:* The amount of tip or gratuity provided by the passenger.

14. **tolls_amount:**
    - *Description:* The total amount of toll charges incurred during the trip.

15. **ehail_fee:**
    - *Description:* A fee associated with the use of electronic hailing services.

16. **improvement_surcharge:**
    - *Description:* An additional charge intended to support taxi industry improvements.

17. **total_amount:**
    - *Description:* The total amount paid by the passenger, including all fare components and additional charges.

18. **payment_type:**
    - *Description:* A numeric code indicating how the passenger paid for the trip (e.g., credit card, cash).

19. **trip_type:**
    - *Description:* A numeric code indicating the type of trip (e.g., street-hail, dispatch).

20. **congestion_surcharge:**
    - *Description:* An additional charge related to congestion pricing.

We are mapping the taxi zones codes in the NYC TLC Trip Record DataFrame (`ft`) with their corresponding names. This involves filtering valid codes, replacing codes with zone names, and creating new columns for pickup and drop-off boroughs.

This code performs the mapping of taxi zone codes with their names, enhancing the readability and interpretability of pickup and drop-off locations in the analysis.

```python
valid_codes = set(tz['LocationID'])

ft = ft[ft['PULocationID'].isin(valid_codes) & ft['DOLocationID'].isin(valid_codes)]

zone_to_borough = tz.set_index('zone')['borough'].to_dict()

ft['PULocationID'] = ft['PULocationID'].replace(tz.set_index('LocationID')['zone'])
ft['DOLocationID'] = ft['DOLocationID'].replace(tz.set_index('LocationID')['zone'])

ft['PUBorough'] = ft['PULocationID'].map(zone_to_borough)
ft['DOBorough'] = ft['DOLocationID'].map(zone_to_borough)
ft.insert(6, 'PUBorough', ft.pop('PUBorough'))
ft.insert(8, 'DOBorough', ft.pop('DOBorough'))
```

We are creating a new DataFrame (`nanval`) to store records from the NYC TLC Trip Record DataFrame (`ft`) that contain NaN values.
```python
nanval = pd.DataFrame(ft[ft[ft.columns].isnull().any(axis=1)])
nanval
```

```python
valratecodes = {1, 2, 3, 4, 5, 6}

ft.loc[~ft['RatecodeID'].isin(valratecodes) & ft['PUBorough'].isin(tz['borough']) & ft['DOBorough'].isin(tz['borough']), 'RatecodeID'] = 1

ft['trip_type'].fillna(1, inplace=True)
```

The provided code aims to handle invalid RatecodeID values in the New York City TLC Taxi dataset. Here's a breakdown of the actions taken:

- Valid RatecodeID values (1, 2, 3, 4, 5, 6) are defined in the set `valratecodes`.
- Rows with invalid RatecodeID, where the RatecodeID is not in the valid set and both Pickup Borough (PUBorough) and Drop-off Borough (DOBorough) are within the valid boroughs, are set to RatecodeID 1.
- Missing values in the '`trip_type`' column are filled with the default value 1.

These adjustments ensure that `RatecodeID` values are within the defined set, contributing to data consistency and accuracy.

We transforms the `lpep_pickup_datetime` and `lpep_dropoff_datetime` columns into the datetime format. Subsequently, it calculates the trip duration in minutes and adds a new column named `trip_duration` to the DataFrame.


```python
# Convert columns to datetime format
ftc['lpep_pickup_datetime'] = pd.to_datetime(ftc['lpep_pickup_datetime'])
ftc['lpep_dropoff_datetime'] = pd.to_datetime(ftc['lpep_dropoff_datetime'])

# Calculate trip duration in minutes
ftc['trip_duration'] = ((ftc['lpep_dropoff_datetime'] - ftc['lpep_pickup_datetime']).dt.total_seconds() / 60).round(2)
```

Filtering out these outliers helps in cleaning the dataset, ensuring that extreme values do not significantly impact the analysis or model performance.

```python
ftc.drop(ftc[ftc['total_amount'] <= 3].index, inplace=True)
ftc.drop(ftc[ftc['trip_duration'] > 720].index, inplace=True)
ftc.drop(ftc[ftc['trip_distance'] > 70].index, inplace=True)
```

The code adds several datetime-related features to the dataset:
- `lpep_pickup_datetime`is converted to the datetime format.
- `day_of_week` is created to represent the day of the week (0=Monday, 1=Tuesday, ..., 6=Sunday).
- `day_type` is derived, categorizing rides as 'weekday' or 'weekend'.
- `day_name` is introduced to indicate the day of the week using day names.
- `overnight` is a binary column indicating if the ride falls within the overnight period (8 pm to 6 am) based on NYC TLC rules.


```python
ftc['lpep_pickup_datetime'] = pd.to_datetime(ftc['lpep_pickup_datetime'])

# Extract day of the week (0=Monday, 1=Tuesday, ..., 6=Sunday)
ftc['day_of_week'] = ftc['lpep_pickup_datetime'].dt.dayofweek

# Create a column indicating the day of the week with day names
day_names = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
ftc['day_type'] = ftc['day_of_week'].apply(lambda x: 'weekday' if x < 5 else 'weekend')
ftc['day_name'] = ftc['lpep_pickup_datetime'].dt.day_name()

# Create a column indicating whether the ride is overnight based on NYC TLC rules
ftc['overnight'] = ftc['lpep_pickup_datetime'].dt.hour.apply(lambda x: 1 if 20 <= x or x < 6 else 0)
```

Transforms the `lpep_pickup_datetime` column into the datetime format. It then categorizes hours into specific time intervals: `Morning` (6 am - 12 pm), `Afternoon` (12 pm - 8 pm), `Evening` (8 pm - 12 am), and `Midnight` (12 am - 6 am). The results are stored in a new column called `time_interval`.


```python
# Convert 'lpep_pickup_datetime' to datetime
ftc['lpep_pickup_datetime'] = pd.to_datetime(ftc['lpep_pickup_datetime'])

# Define time intervals
morning_hours = range(6, 12)
afternoon_hours = range(12, 20)
evening_hours = range(20, 24)
midnight_hours = range(0, 6)

# Function to categorize hours
def categorize_hour(hour):
    if hour in morning_hours:
        return 'Morning'
    elif hour in afternoon_hours:
        return 'Afternoon'
    elif hour in evening_hours:
        return 'Evening'
    elif hour in midnight_hours:
        return 'Midnight'

# Apply categorization to create a new column 'time_interval'
ftc['time_interval'] = ftc['lpep_pickup_datetime'].dt.hour.apply(categorize_hour)
```
